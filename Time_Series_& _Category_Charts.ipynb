### Imports & Data Loading

```python
# Data Manipualation
import pandas as pd

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inline
%matplotlib inline
```

### Data Preparation (Minimal Cleaning)

```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Create Month-Year column for aggregation
df['Month_Year'] = df['Order Date'].dt.to_period('M').astype(str)
```

### Time Series Analysis – Sales Over Time (LINE CHART)

### Monthly Sales Trend

```python
monthly_sales = (
    df.groupby('Month_Year')['Sales']
    .sum()
    .reset_index()
)

plt.figure(figsize=(14,6))
plt.plot(monthly_sales['Month_Year'], monthly_sales['Sales'])
plt.title("Monthly Sales Trend Over Time")
plt.xlabel("Month-Year")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Why a line chart?

* Line charts are ideal for **time-based trends**
* They clearly show **growth, decline, and seasonality**
* Easy to interpret changes over time

### Category Comparison – BAR CHART

### Sales by Category

```python
category_sales = (
    df.groupby('Category')['Sales']
    .sum()
    .reset_index()
)

plt.figure(figsize=(8,5))
sns.barplot(data=category_sales, x='Category', y='Sales')
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

### Why a bar chart?

* Best for **comparing values across categories**
* Makes it easy to identify the **highest and lowest performing categories**
* Clear axis labeling improves readability

### Category Share – PIE CHART

### Percentage Share of Sales by Category

```python
plt.figure(figsize=(7,7))
plt.pie(
    category_sales['Sales'],
    labels=category_sales['Category'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Sales Share by Category")
plt.show()
```

### Why a pie chart?

* Useful for showing **proportions and contribution**
* Helps answer: *Which category contributes the most to total sales?*
* Limited categories make the chart readable and effective
